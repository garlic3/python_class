```python
# 동적 길이 스택
# 1. create(size): 스택을 생성(초기화)
# 2. is_empty(): 스택이 비어있다면 true, 아니라면 false
# 3. is_full(): 스택이 다 차있다면 true, 아니라면 false
# 4. push(): 스택에 데이터를 넣기
# 5. pop(): 스택에서 데이터를 꺼내기
# 6. peek(): top에 위치한 데이터를 반환

class MyStack:
    
    # 빈 스택을 생성
    def __init__(self):
        self.container = []
    
    def is_empty():
        return len(self.container) <= 0
    
    def push(self, data):
        self.container.append(data)
        
    def pop(self):
        self.container.pop()
    
    def peek(self):
        return self.container[-1]
    
    def size(self):
        return len(self.container)
    
    def dump(self):
        print(self.container[:])
    
    def clear(self):
        self.container = []
```

```python
# 정적 길이 스택
# 1. create(size): 스택을 생성(초기화)
# 2. is_empty(): 스택이 비어있다면 true, 아니라면 false
# 3. is_full(): 스택이 다 차있다면 true, 아니라면 false
# 4. push(): 스택에 데이터를 넣기
# 5. pop(): 스택에서 데이터를 꺼내기
# 6. peek(): top에 위치한 데이터를 반환

class MyStack2:
    
    def __init__(self, size):
        
        self.size = size
        self.container = [None]*size
        self.pointer = 0
        
    def is_empty(self):
        return self.pointer <= 0
    
    def is_full(self):
        return self.pointer >= self.size
    
    def pop(self):
        if self.is_empty():
            return print("빈 스택입니다")
        self.pointer -= 1
        return self.container[self.pointer]
    
    def push(self, data):
        if self.is_full():
            return print("가득찬 스택입니다")
        self.container[self.pointer] = data
        self.pointer += 1
    
    def peek(self):
        if self.is_empty():
            return print("빈 스택입니다")
        return self.container[self.pointer-1]
    
    def clear(self):
        self.pointer = 0
        self.container[:] = None
        
    def dump(self):
        if self.is_empty():
            return print("빈 스택입니다")
        print(self.container[:self.pointer])
    
    
m = MyStack2(3)
m.peek()
m.pop()
m.dump()

m.push(1)
m.push(2)
m.push(3)
m.push(4)
```

```python
# 스택 활용 예제
# 회문을 검사하는 사용자 정의 함수 만들기(위에 만들어놓은 스택을 활용)
# 사용자로부터 문자열을 입력받음
# 회문일 경우 '회문입니다' 라고 출력
# 회문이 아닐경우 '회문이 아닙니다'라고 출력

def check_palindrome(text):
    m = MyStack2(len(text))
    for i in text:
        m.push(i)
        
    for i in text:
        if i == m.peek():
            m.pop()
        else:
            return print("회문이 아닙니다")
    if m.is_empty():
        return print("회문입니다")

check_palindrome('aba')
    
```

```python
# 동적 큐
# 1. create(size): 큐를 생성(초기화)
# 2. is_empty(): 비어있다면 true, 아니라면 false
# 3. is_full(): 가득 차 있다면 true, 아니라면 false
# 4. enqueue(): 큐에 데이터를 추가
# 5. dequeue(): 큐에서 데이터를 꺼내기
# 6. peek(): 맨 처음 데이터를 반환(디큐에서 꺼내질 데이터)

class MyQueue:
    def __init__(self):
        self.container = []
        
    def is_empty(self):
        return len(self.container) <= 0
    
    def enqueue(self, data):
        self.container.append(data)
        
    def dequeue(self):
        self.container.pop(0)
    
    def peek(self):
        return self.container[0]
    
    def size(self):
        return len(self.container)
    
    def dump(self):
        print(self.container[:])
    
    def clear(self):
        self.container = []
```

```python
# 정적 큐

class MyQueue2:
    
    def __init__(self, size):
        self.size = size
        self.container = [None]*size
        self.pointer = 0
        
    def is_empty(self):
        return self.pointer <= 0
    
    def is_full(self):
        return self.pointer >= self.size
    
    def enqueue(self, data):
        self.container[self.pointer] = data
        self.pointer += 1
    
    def dequeue(self):
        self.container[:] = self.container[1:]
        self.pointer -= 1
    
    def peek(self):
        return self.container[0]
    
    def dump(self):
        print(list(self.container[:self.pointer]))
        
    def size(self):
        return self.pointer
    
    def clear(self):
        self.container[:] = None
```

```python
# 링 버퍼 큐


class RingQueue:
    
    def __init__(self, size):
        
        self.pointer = 0
        self.front = 0
        self.rear = 0
        self.size = size
        self.container = self.size * [None]
        
    def is_empty(self):
        return self.pointer <= 0
    
    def is_full(self):
        return self.pointer >= self.size
    
    
    def enqueue(self, data):
        self.container[self.rear] = data
        self.rear += 1
        if self.rear == self.size:
            self.rear = 0
        self.pointer += 1

    def dequeue(self):
        data = self.container[self.front]
        self.front += 1
        self.pointer -= 1
        if self.front == self.size:
            self.front == 0
            
        return data
    
    def peek(self):
        return self.container[self.front]
    
    def clear(self):
        self.pointer = 0
        self.front = 0
        self.rear = 0
        
    def dump(self):
        print(self.container[:self.rear-2])
    
    
```

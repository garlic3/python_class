```python
import requests
from bs4 import BeautifulSoup as bs 

url = 'http://naver.com'

response = requests.get(url)
# status_code 상태코드
print(response.status_code)

# text 웹 페이지 소스 -> 보기 쉽게
print(response.text)

# content 웹 페이지 소스 -> 그대로 
print(response.content)
```

```python
url = 'http://naver.com'
html = requests.get(url)
soup = bs(html.text, 'html.parser')
print(soup)
```

```python
# find 예제

# 태그
strong = soup.find('strong')
print(strong)

# 태그내용
strong = soup.find('strong').text
print(strong)

# 속성값 이용
# <ul class="kwd_lst _recent">
strong = soup.find('ul').get('class')
print(strong)
```

```python
# find_all 예제
# 태그
f_list = soup.find_all('strong')
print(f_list)
print(len(f_list))

# 태그내용
strong = soup.find_all('strong')[0]
print(strong)
strong = soup.find_all('strong')[0].text
print(strong)

# 속성값
a_list = soup.find_all('li')[0].get('class')
print(a_list)
```

```python
# class 예제
# find()
strong = soup.find('strong', class_='title')
print(strong)

# # print(strong.text)
strong = soup.find('strong', class_='title').text
print(strong)

# find_all()
f_list = soup.find_all('li', class_='nav_item')
print(len(f_list))
for i in f_list:
    print(i.text)
```

```python
# id 예제 
# id는 특정값이므로 find_all 사용 x
div = soup.find('div', id='NM_FAVORITE')
print(div)
print(div.text)
```

```python
# 네이버 영화
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver'
html = requests.get(url)
soup = bs(html.text, 'html.parser')
print(soup)
```

```python
# 평점 박스 찾기
movie_box = soup.find('table', class_='list_ranking')
# print(movie_box)

movie = movie_box.find_all('div', class_='tit3')
# print(movie)
# print(movie[0])
print(type(movie))

for i in movie:
  # print(i.find('a'))
  # print(i.find('a').text)
    print(i.find('a').get('title'))

movie_list = []
rank = 1

for i in movie:
    m = i.find('a').get('title')
    movie_list.append(f"{rank}위: {m}")
    rank +=1

print(movie_list)  
```

```python
# 실습2
# 영화 인기 검색어, 영화인 인기검색어, 박스오피스 크롤링해서 리스트로 담은 후 출력
lanking_box = soup.find_all('ul', class_='r_ranking')
# print(lanking_box[0])
movies = lanking_box[0].find_all('a')
print(movies)

# 영화 인기 검색어
movie_search_rank = []
rank = 1
for i in movies:
    movie = i.get('title')
    movie_search_rank.append(f"{rank}위 : {movie}")
    rank += 1

print(movie_search_rank)

# 영화인 인기검색어
actor_search_rank = []
actors = lanking_box[1].find_all('a')
rank = 1
for i in actors:
    actor = i.get('title')
    actor_search_rank.append(f"{rank}위 : {actor}")
    rank += 1

print(actor_search_rank)

# 박스오피스
box_office_rank = []
boxes = lanking_box[2].find_all('a')
rank = 1
for i in boxes:
    box = i.get('title')
    box_office_rank.append(f"{rank}위 : {box}")
    rank += 1
print(box_office_rank)
```

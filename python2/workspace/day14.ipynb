```python
import random 
# 수학, 영어, 국어, 사회, 미술
# 학생 300명
school = []
for i in range(50):
    gender = random.randint(0,1)
    if gender == 0:
        gender = 'f'
    else:
        gender = 'm'
    math, eng, kor, his, art = random.sample(range(1,101), 5)
    school.append([gender, math, eng, kor, his, art])        
# print(school)

df = pd.DataFrame(school)
df.columns = ['gender', 'math', 'eng', 'kor', 'his', 'art']
# 열 이름 일부만 변경
# df.rename(columns={'before':'after'})
# df.rename(columns={'gender':'gen'}) 
# 변경 후 새로 담아줘야함

# 인덱스 이름 변경
# 
df.index = df.index+1
df
```

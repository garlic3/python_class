```python
class Node:
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next
        
class LinkedList:
    # 초기화
    def __init__(self):
        # 머리노드
        self.head = None
        # 노드 개수
        self.node_num = 0

    def insert_first(self, data):
        if self.is_empty():
            # 노드 추가
            self.head = Node(data)
            self.node_num += 1
            # 기존 노드가 있는 경우
        else:
            prev = self.head
            self.head = Node(data, prev)
            self.node_num += 1
            
    def insert(self, i, data):
        if self.is_empty():
            return print("빈 노드 입니다. 인덱스 에러")
        elif i > self.node_num -1:
            return print("인덱스 에러")
        else:
            prev = self.get_node(i-1)
            prev.next = Node(data, prev.next)
            self.node_num += 1
    
    def insert_last(self, data):
        if self.is_empty():
            self.head = Node(data)
            self.node_num += 1
        else:
            prev = self.get_node(self.node_num-2)
            prev.next = Node(data)
            self.node_num += 1
            
    def remove_first(self):
        if self.is_empty():
            return print("빈 노드입니다")
        else:
            prev = self.head
            self.head = prev.next
            self.node_num -= 1
    
    def remove(self, i):
        if self.is_empty():
            return print("빈 노드입니다")
        else:
            prev = self.get_node(i-1)
            current = prev.next
            prev.next = current.next
            self.node_num -=1 
            
    def remove_last(self):
        if self.is_empty():
            return print("빈 노드입니다")
        else:
            prev = self.get_node(self.node_num-2)
            prev.next = None
            self.node_num -=1

    # i번째 노드 찾기
    def get_node(self, i):
        current = self.head
        for num in range(i):
            current = current.next
        return current
    
    # 모든 노드 삭제
    def clear(self):
        self.head = None
        self.node_num = 0
    
    # 모든 노드 출력
    def get_all(self):
        if self.is_empty():
            return print("빈 노드입니다")
        else:
            current = self.head
            for i in range(self.node_num):
                print(f"{i}번째 데이터 = {current.data}")
                current = current.next
            
    # 노드 인덱스 확인(첫번째 만나는 데이터)
    def get_idx(self, data):
        current = self.head
        for i in range(self.node_num):
            if(current.data == data):
                idx = i
                return idx
            current = current.next
            
    # 데이터 개수 확인
    def get_data_num(self, data):
        current = self.head
        count = 0
        for i in range(self.node_num):
            if(current.data == data):
                count += 1
            current = current.next
        return count
    
    
    # 개수 확인        
    def get_size(self):
        return self.node_num
        
    # 비어 있는지 확인    
    def is_empty(self):
        return self.node_num <= 0

```

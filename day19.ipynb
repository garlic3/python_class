```python
class Node:
    def __init__(self, key, value, left=None, right=None):
        self.key = key
        self.value = value
        self.left = left
        self.right = right
        
        
class BinarySearchTree:
    def __ini__(self):
        self.root = None
    
    # 키값으로 노드를 검색
    def search(self, key):
        current = self.root
        while True:
            if current is None:
                return print("해당 키가 존재하지 않습니다")
            if key == current.key:
                return current.value
            elif key < current.key:
                current = current.left
            else:
                current = current.right
    

    def insert(self, key, value):
        current = self.root
        # 삽입할때 경우의 수 3개
        # 1. 빈 트리 2. root보다 작은값 3. root보다 큰값
        if self.is_empty():
            self.root = Node(key, value)
            return
        elif key < current.key:
            if current.left is None:
                current.left = Node(key, value)
            else:
                self.insert(key, value)
        else:
            if current.right is None:
                current.right = Node(key, value)
            else:
                self.insert(key, value)
            
    def is_empty(self):
        return self.root == None
    
    
    def remove(self, key):
        
        # 스캔중인 노드
        current = self.root
        # 스캔중인 노드의 부모 노드
        parent = None
        # 스캔중인 노드가 부모의 왼쪽 노드인지 체크
        is_left_child = True
        
        # 삭제할 키 위치 찾기
        while True:
            #키가 존재하지 않음
            if current is None:
                return print("키가 트리에 존재하지 않습니다")
            #키를 찾음
            if key == current.key:
                break    
            else:
                # 가지를 내려가기전에 부모를 설정
                parent = current
                # 만약 키가 루트보다 작으면 왼쪽으로
                if key < current.key:
                    # 왼쪽자식으로 내려감
                    is_left_child = True
                    # 왼쪽 서브트리에서 검색
                    current = current.left
                # 만약 키가 루트보다 크면 오른쪽으로
                else:
                    # 키가 루트보다 크면 오른쪽으로
                    is_left_child = False
                    # 오른쪽 서브트리에서 검색
                    current = current.right
        
        # 삭제 노드에 왼쪽 자식이 없으면 왼쪽 포인터가 None
        # 삭제 노드에 오른쪽 자식이 없으면 오른쪽 포인터가 None
        
        # 왼쪽 자식이 없으면
        if current.left is None:
            # 기준노드가 루트라면 새로운 루트노트 설정
            if current is self.root:
                self.root = current.right
            # 왼쪽 서브트리에 있다면
            elif is_left_child:
                # 부모의 왼쪽 포인터가 오른쪽 자식을 가리킴
                parent.left = current.right
                # 부모의 오른쪽 포인터가 오른쪽 자식을 가리킴
            else:
                parent.right = current.right
        # 오른쪽 자식이 없으면        
        elif current.right is None:
            # 기존노드가 루트라면 새로운 루트노트 설정
            if current is self.root:
                self.root = current.left
            elif is_left_child:
                # 부모의 왼쪽 포인터가 나의 왼쪽 자식을 가리킴
                parent.left = current.left
            else:
                # 부모의 오른쪽 포인터가 나의 왼쪽 자식을 가리킴
                parent.right = current.right
    
        else:
            # 자식 노드가 2개인 경우
            parent = current
            # 서브트리 안에서 가장 큰 노드
            left = current.left
            is_left_child = True
            # 서브트리 안에서 가장 큰 노드 left 찾기 
            while left.right is not None:
                parent = left
                left = left.right
                is_left_child = False
            
            # left의 키를 current로 이동
            current.key = left.key
            # left의 데이터를 p로 이동
            current.value = left.value
            
            # left 삭제 
            if is_left_child:
                parent.left = left.left
            # left 삭제
            else:
                parent.right = left.left
                        
        return True
    
    # 가장 작은 키를 찾아 반환
    def min_key(self):
        if self.root is None:
            return None
        p = self.root
        while p.left is not None:
            p = p.left
        return p.key
    
    # 가장 최대 값을 찾아 반환
    def max_key(self):
        if self.root is None:
            return None
        p = self.root
        while p.right is not None:
            p = p.right
        return p.key
```
